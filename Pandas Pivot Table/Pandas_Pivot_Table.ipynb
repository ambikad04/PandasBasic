# **Pandas Pivot Table**

```python
import pandas as pd
import numpy as np
```

```python
food = pd.read_csv('food.csv')
food.head()
```

```python
food.pivot_table(index=['City','Gender'], columns=['Item','Frequency'], values='Spends', aggfunc='mean')
```
